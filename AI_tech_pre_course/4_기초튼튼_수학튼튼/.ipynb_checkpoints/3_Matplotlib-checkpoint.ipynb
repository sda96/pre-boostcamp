### matplotlib

- pyplot 객체를 사용하여 데이터를 표시
- pyplot 객체에 그래프를 쌓은 다음 show로 flush(보여줌)
- matplotlib는 argument를 kwargs로 받음
- 고정된 argument가 없어서 alt+tab으로 확인이 어려움
- 공홈에가서 예제 확인

pyplot이라는 객체안에 figure라는 모판(도화지), 객체를 생성함  
pyplot 객체 사용시, 기본 figure 에 그래프가 그려짐

```python
import matplotlib.pyplot as plt

x = range(100)
y = range(100)
plt.plot(x,y)
plt.show()
```

plt.show()를 사용하기 전까지는 figure 객체에 plot들이 그려지다가 plt.show()를 사용하면 figure 객체를 비우면서 화면에 flush를 해줌

```python
import numpy as np

x_1 = range(100)
y_1 = [np.cos(value) for value in x_1]

x_2 = range(100)
y_2 = [np.sin(value) for value in x_2]

plt.plot(x_1, y_1)
plt.plot(x_2, y_2)
plt.show()
```

기본 figure() 그림판에 좀 더 작은 크기의 그림판을 만들 수 있음

```python
fig = plt.figure() # figure 반환
fig.set_size_inches(10,5) # 크기 지정
ax_1 = fig.add_subplot(1,2,1) # 1행 2열의 1열 figure
ax_2 = fig.add_subplot(1,2,2) # 1행 2열의 2열 figure

ax_1.plot(x_1, y_1, c="b")
ax_2.plot(x_2, y_2, c="g")
plt.show() # show & flush
```

#### set color

- color 속성을 사용
- float > 흑백, rgb, predefined color 사용

```python
plt.plot(x_1, y_1, color = "#eeefff")
plt.plot(x_1, y_1, color = "r")
plt.show()
```

#### set linestyle

- ls 또는 linestyle 속성 사용

```python
plt.plot(x_1, y_1, c = "b", linestyle = "dashed")
plt.plot(x_2, y_2, c = "r", ls = "dotted")
plt.show()
```

#### set title & legend(범례)

- pyplot에 title 함수를 사용하며, figurㄷ의 subplot별 입력 가능
- Legend 함수로 범례를 표시하며 loc parameter를 이용한 위치 지정도 가능

```python
plt.plot(x_1, y_1, c = "b", linestyle = "dashed", label = "line_1")
plt.plot(x_2, y_2, c = "r", ls = "dotted", label = "line_2")
plt.title("Two lines, $y = f(x)$") # Latex도 사용 가능
plt.legend(shadow=True, fancybox=True, loc="lower right")
plt.show()
```

### Scatter

- scatter 함수 사용, marker: scatter 모양 지정
- s: 데이터의 크기를 지정 가능함

```python
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)

plt.scatter(data_1[:,0], data_1[:,1], c = "b", marker="x")
plt.scatter(data_2[:,0], data_2[:,1], c = "r", marker="*")
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

#### bar chart

```python
data = [
    [5., 25. , 50. ,20.],
    [4., 23., 51., 17.],
    [6., 22., 52., 19]
       ]
x = np.arange(4)
plt.bar(x + 0.00, data[0], color="b", width = 0.25)
plt.bar(x + 0.25, data[1], color="r", width = 0.25)
plt.bar(x + 0.50, data[2], color="g", width = 0.25)
plt.xticks(x + 0.25, ("A", "B", "C", "D"))
plt.show()
```

```python
data = np.array(data)
color_list= ["b", "g", "r"]
data_label = ["A", "B", "C"]
x = np.arange(data.shape[1])
for i in range(data.shape[0]):
    plt.bar(x, data[i], bottom= np.sum(data[:1], axis=0),
           color = color_list, label= data_label[i])
plt.legend()
plt.show()
```

#### histogram

```python
x = np.random.randn(1000)
plt.hist(x, bins=100) # 1000개 데이터를 100개 로 나누어라
plt.show()
```

#### box plot

```python
plt.boxplot(x)
plt.show()
```
